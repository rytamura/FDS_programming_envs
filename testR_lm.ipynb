```r
# サンプルデータの作成
x <- 1:10
y <- 2 * x + rnorm(10)

# データフレームにまとめる
df <- data.frame(x = x, y = y)

# 線形回帰モデルの作成
model <- lm(y ~ x, data = df)

# 回帰結果の表示
summary(model)

# macOS用の日本語フォント指定（警告が出ない例）
par(family="Meiryo") # デフォルトフォント（日本語が表示されていればこれでOK）

# 散布図と回帰直線の描画
plot(df$x, df$y, main = "線形回帰デモ", xlab = "x", ylab = "y")
abline(model, col = "red")
```
