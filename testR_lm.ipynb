```r
# サンプルデータの作成
set.seed(1223) # 再現性のためのシード設定
nrepl = 20000

coefs = list()

for (i in 1:nrepl) {
  x <- rnorm(100)
  y <- 2 * x + rnorm(100)
  
  # データフレームにまとめる
  df <- data.frame(x = x, y = y)
  
  # 線形回帰モデルの作成
  model <- lm(y ~ x, data = df)
  
  # 回帰結果の表示
  coefs[[i]] = coef(model)
}
```

```r
coef = do.call(rbind, coefs)
colnames(coef) = c("intercept", "slope")
```

```r
par(family= "HiraKakuProN-W3")
suppressWarnings( 
  hist(coef[,1], breaks="scott", main="切片推定値の分布", xlab="切片", cex=2)
)
```
